```python
import yt
import numpy as np

# Создаем тестовые данные (3D массив плотности)
data = np.random.random((64, 64, 64))

# Описываем данные для yt
data_dict = dict(density = (data, "g/cm**3"))
bbox = np.array([[-1.5, 1.5], [-1.5, 1.5], [-1.5, 1.5]]) # границы области

# Загружаем как равномерную сетку
ds = yt.load_uniform_grid(data_dict, data.shape, length_unit="pc", bbox=bbox)
```

```python
# 1. Создаем сцену
sc = yt.create_scene(ds, field='density')

# 2. Настраиваем камеру (опционально)
cam = sc.camera
cam.resolution = (800, 800)
cam.position = ds.arr([3.0, 3.0, 3.0], 'pc') # отодвигаем камеру

# 3. Визуализируем
sc.show()
```
